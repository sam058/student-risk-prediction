```python
# identifying columns with zero values and further dropping them 

# Separate feature columns explicitly
feature_cols = clustered.drop(columns=["cluster", "cluster_label"], errors="ignore")

# Identify columns that are zero for all students
zero_cols = feature_cols.columns[(feature_cols == 0).all()]

zero_cols

```

```python
clustered_clean = clustered.drop(columns=zero_cols)

clustered_clean.head()
```

```python
numeric_features = clustered_clean.select_dtypes(include="number")

numeric_features = numeric_features.drop(columns=["cluster"], errors="ignore")

usage_counts = (numeric_features > 0).sum()
usage_counts.sort_values()


```

```python
threshold = 0.01 * len(numeric_features)

low_usage_cols = usage_counts[usage_counts < threshold].index

low_usage_cols
```

```python
clustered_final = clustered_clean.drop(columns=low_usage_cols)

clustered_final.head()
```

```python
student_info = pd.read_csv("data/studentInfo.csv")

labels = student_info[["id_student", "final_result"]].copy()

labels["target"] = labels["final_result"].apply(
    lambda x: 0 if x in ["Pass", "Distinction"] else 1
)

labels["target"].value_counts()
```

```python
final_dataset = clustered.merge(
    labels[["id_student", "target"]],
    on="id_student",
    how="inner"
)

final_dataset.shape
```

```python

final_dataset["target"].value_counts()
```

```python
final_dataset.shape
```

```python
behavior_cols = [
    col for col in final_dataset.columns
    if col not in ["id_student", "cluster", "cluster_label", "target"]
]

final_dataset_active = final_dataset[
    final_dataset[behavior_cols].sum(axis=1) > 0
]

final_dataset_active.shape
final_dataset_active["target"].value_counts()
```

```python
final_dataset_active.to_csv(
    "outputs/final_prediction_dataset.csv",
    index=False
)
```
